```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text = "All work and no play makes Jack a dull boy."
doc = nlp(text)
```

```python
for token in doc:
    print(token.text,
          token.lemma_, # 표제어(사전에 등록되는 형태)
          token.pos_, # 단어의 품사
          token.tag_, # 자세한 품사
          token.is_stop) # 불용어 여부
```

```python
spacy.explain('PROPN')
```

```python
for token in doc:
  print(token.text,
      token.dep_, # 의존 관계
      token.head) # 지배소
```

#Python에서 한국어 토큰화 방법   
• kiwi: C++로 개발된 형태소 분석기   
• spaCy: 딥러닝 기반 형태소 분석기. 의존 구문 분석도 제공.

```python
# 설치
!pip install kiwipiepy
# 불러오기
from kiwipiepy import Kiwi
kiwi = Kiwi()
```

```python
text = '오늘은 자연어 처리를 배우기 좋은 날이다.'
result = kiwi.tokenize(text)
result
```

```python
# kiwi 명사 추출 함수
def extract_nouns(text):
    result = kiwi.tokenize(text)
    for token in result:
        if token.tag in ['NNG', 'NNP']:
            yield token.form
```

```python
list(extract_nouns('어제는 홍차를 마시고, 오늘은 커피를 마셨다.'))
```
